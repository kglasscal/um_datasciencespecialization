## Question 1
Write a function called `proportion_of_education` which returns the proportion of children in the dataset who had a mother with the education levels equal to less than high school (<12), high school (12), more than high school but not a college graduate (>12) and college degree.

*This function should return a dictionary in the form of (use the correct numbers, do not round numbers):* 
```
    {"less than high school":0.2,
    "high school":0.4,
    "more than high school but not college":0.2,
    "college":0.2}
```


````python
import pandas as pd
import numpy as np

def proportion_of_education():

    # Maternal education 
    # 1: <12 years, 
    # 2:12 years, 
    # 3: >12 years, not a college graduate
    # 4: College graduate
    columns = ['EDUC1']
    education_df = pd.read_csv("NISPUF17.csv", usecols=columns)
    df_len = len(education_df)
    data = education_df.value_counts()
    
    return {"less than high school": data.get(1.0)/df_len, "high school": data.get(2.0)/df_len, "more than high school but not college": data.get(3.0)/df_len, "college": data.get(4.0)/df_len}

    raise NotImplementedError()
````

````python
def chickenpox_by_sex():
    # YOUR CODE HERE
    columns                 = ['SEX', 'HAD_CPOX', 'P_NUMVRC']
    vaccine_data_df         = pd.read_csv("NISPUF17.csv", usecols=columns)
    df_len                  = len(vaccine_data_df)

    # get entries with 'P_NUMVRC' = 1, we only need vaccinated individuals
    vaccinated_df     = vaccine_data_df[vaccine_data_df['P_NUMVRC'] > 0]
    
    gender_df         = vaccinated_df[vaccinated_df['SEX'] <= 2]
    had_pox_df        = vaccinated_df[vaccinated_df['HAD_CPOX'] <= 2]
    gender_had_pox_df = had_pox_df[had_pox_df['SEX'] <= 2] 

    # Series containing 
    vaccinated_data = vaccine_data_df.value_counts()
    gender_data = gender_df.value_counts()
    had_pox_data = had_pox_df.value_counts()

    boy_vaccinated_without_cpox  = 0     # 1, 1, > 0
    boy_vaccinated_with_cpox     = 0     # 1, 2, > 0

    girl_vaccinated_without_cpox = 0     # 2, 1, > 0
    girl_vaccinated_with_cpox    = 0     # 2, 2, > 0

    for gender in range(1, 3) :
        for had_pox in range(1, 3) :
            for n_vaccines in range(1, len(vaccinated_data)+1) :
                try :
                    had_pox_data[had_pox, gender, n_vaccines]
                except KeyError :
                    continue

                if gender == 1 :
                    if had_pox == 1 :
                        boy_vaccinated_without_cpox += had_pox_data[had_pox, gender, n_vaccines]
                    else :
                        boy_vaccinated_with_cpox += had_pox_data[had_pox, gender, n_vaccines]
                else:
                    if had_pox == 1 :
                        girl_vaccinated_without_cpox += had_pox_data[had_pox, gender, n_vaccines]
                    else :
                        girl_vaccinated_with_cpox += had_pox_data[had_pox, gender, n_vaccines]

    print ("boy_vaccinated_with_cpox", boy_vaccinated_with_cpox)
    print ("boy_vaccinated_without_cpox", boy_vaccinated_without_cpox)
    print ("girl_vaccinated_without_cpox", girl_vaccinated_without_cpox)
    print ("girl_vaccinated_with_cpox", girl_vaccinated_with_cpox)
    if boy_vaccinated_with_cpox == 0 :
        raise NotImplementedError()
    if girl_vaccinated_with_cpox == 0 :
        raise NotImplementedError()

    return {"male": boy_vaccinated_without_cpox/boy_vaccinated_with_cpox, "female": girl_vaccinated_without_cpox/girl_vaccinated_with_cpox}

    raise NotImplementedError()
````
