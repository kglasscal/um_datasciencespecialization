```python
def CitiesClean(cities) :
#    print('===================== CitiesClean ======================')
    GeneralClean(cities)

    cities.replace('[—]', "", regex=True, inplace=True)
   
    print("CitiesClean\n", cities.head(5))
    columnTitles = ['Metropolitan area','Population', 'NFL', 'MLB', 'NBA', 'NHL']
    cities.columns = columnTitles
    print("Set column\n", cities.head(5))
    cities['Population'].astype(float)
    cities.set_index('Metropolitan area', inplace=True)
    print("Complete\n", cities.head(5))
    
    return cities
```

```python
def NHLClean(Sport) :
    GeneralClean(Sport)
    pattern = re.compile( '.* Division')
    
    Sport = Sport[['team','W', 'L']].    \
        loc[(Sport['year'] == 2018)]
    Sport.columns = ['Team','W', 'L']

    atlantic = Sport["Team"] == "Atlantic Division"
    metropolitan = Sport["Team"] == "Metropolitan Division"
    central = Sport["Team"] == "Central Division"
    pacific = Sport["Team"] == "Pacific Division"
    divisions = atlantic | metropolitan | central | pacific

    Sport.drop(Sport[divisions].index, inplace=True)
    Sport.index = pd.RangeIndex(len(Sport.index))

    Sport['W'] = Sport['W'].astype(int)
    Sport['L'] = Sport['L'].astype(int)

    Sport.set_index('Team', inplace=True)

    return Sport
```

```python
def combineMetroAndSport(CityData, NHLData, Sport) :
    # Table
    NHLCombinedTable = pd.DataFrame(columns=['Metro Area', 'Population', 'W', 'L'])

    for regionDict in MasterNameDictionary :
        region      = regionDict['region']
        population  = CityData.at[regionDict['region'], 'Population']
        wins        = 0
        losses      = 0

        # accumulate wins and losses
        if Sport in regionDict :
            if type(regionDict[Sport]) == list :
                for i in range (len(regionDict[Sport])) :
                    wins   =  wins   + NHLData.at[regionDict[Sport][i], 'W']
                    losses =  losses + NHLData.at[regionDict[Sport][i], 'L']
            else :
                wins   = wins   +  NHLData.at[regionDict[Sport], 'W']
                losses = losses +  NHLData.at[regionDict[Sport], 'L']

            dict = {'Metro Area': region, 'Population' : population, 'W' : wins, 'L' : losses}
            row = pd.Series(dict)

            NHLCombinedTable = NHLCombinedTable.append(row, ignore_index=True)

    return NHLCombinedTable
```
