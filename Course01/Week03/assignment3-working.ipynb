## def getEnergyIndicators(excelName)

````python
def getEnergyIndicators() :

    # Requirements 1, 2, 5
    energyIndicators   = pd.read_excel(
        io             = 'assets/Energy Indicators.xls',
        sheet_name     = 'Energy',
        na_values      = '...',
        usecols        = 'C:F',
        skiprows       = range(17),
        nrows          = 247-20
    )
    # Requirements 3, 5
    energyIndicators.columns    = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

    # Requirements 7
    energyIndicators['Country'] = energyIndicators['Country'].str.replace('\d+', '',case=True, regex=True)
    energyIndicators['Country'] = energyIndicators['Country'].str.replace(r' \([^)]*\)', '', regex=True)
    energyIndicators['Country'] = energyIndicators['Country'].where(energyIndicators['Country'].notna(), np.NaN)

    # Set index to the country column
    energyIndicators.set_index('Country', inplace=True)

    # Requirements 6
    energyIndicators.rename(index={
        'Republic of Korea': 'South Korea', 
        'United States of America': 'United States',
        'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
        'China, Hong Kong Special Administrative Region': 'Hong Kong'
    }, inplace=True)

    # Requirements 4
    energyIndicators['Energy Supply'] = energyIndicators['Energy Supply'].apply(lambda x: x * 1e6)

    return energyIndicators

    raise NotImplementedError()
````

````python
def getGDP() :
    # Read 'Country Name', 2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', 2015'
    
    colList = ['Country Name', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    gdp = pd.read_csv(
        'assets/world_bank.csv', 
        header = 4,
        usecols=colList
    )
    gdp.columns = ['Country', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']

    gdp_obj = gdp.select_dtypes(['object'])
    gdp[gdp_obj.columns] = gdp_obj.apply(lambda x: x.str.strip())

    gdp.set_index('Country', inplace=True)

    gdp.rename(index={
        'Korea, Rep.': 'South Korea', 
        'Iran, Islamic Rep.': 'Iran',
        'Hong Kong SAR, China': 'Hong Kong'
    }, inplace=True)

    return gdp

    raise NotImplementedError()
````

````python
def getScienceRank() :
    scienceRank = pd.read_excel(
        'assets/scimagojr-3.xlsx',
        sheet_name='Sheet1',
        usecols='A:H'
    )
    scienceRank_obj = scienceRank.select_dtypes(['object'])
    scienceRank[scienceRank_obj.columns] = scienceRank_obj.apply(lambda x: x.str.strip())
    scienceRank['Country'].apply(lambda val: unicodedata.normalize('NFD', val).encode('ascii', 'ignore').decode())

    scienceRank.columns = ['Rank', 'Country', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 'Citations per document', 'H index']
    scienceRank.set_index('Country', inplace=True)

    return scienceRank

    raise NotImplementedError()
````

### Question 10
Create a new column with a 1 if the country's % Renewable value is at or above the median for all countries in the top 15, and a 0 if the country's % Renewable value is below the median.

*This function should return a series named `HighRenew` whose index is the country name sorted in ascending order of rank.*

````python
def answer_ten():
    renewableDF                 = loadAnalysisFrame()
    renewableDF['Renewable']    = renewableDF['% Renewable']
    renewableMedian             = renewableDF['% Renewable'].median(axis=0, skipna=True)

    renewableDF.loc[(renewableDF['% Renewable'] >= renewableMedian),'Renewable'] = 1
    renewableDF.loc[(renewableDF['% Renewable'] < renewableMedian),'Renewable'] = 0
    renewableDF.sort_values(by=['Rank'], axis=0, ascending=True, inplace=True, kind='quicksort', na_position='last')
    renewableDF['Renewable'] = renewableDF['Renewable'].astype(int)

    HighRenew = renewableDF['Renewable']
#    print(HighRenew.index)

    return HighRenew

    raise NotImplementedError()
````

````python
# Target: a dataframe containing a list of continents and an index
#         a list of countries on the continent
#         a list of country populations on the continent
#         a list of the number of countries/continent
#         a list of sum, mean, and std for each continent
#     
# Create a ContinentData dataframe consisting of the following lists
#     1) List of continents
#     2) A list of lists of countries
#     3) A list of lists of the population in each country
def answer_eleven():
    ContinentDict  = {
        'China':'Asia', 
        'United States':'North America', 
        'Japan':'Asia', 
        'United Kingdom':'Europe', 
        'Russian Federation':'Europe', 
        'Canada':'North America', 
        'Germany':'Europe', 
        'India':'Asia',
        'France':'Europe', 
        'South Korea':'Asia', 
        'Italy':'Europe', 
        'Spain':'Europe', 
        'Iran':'Asia',
        'Australia':'Australia', 
        'Brazil':'South America'
    }
    

    # create ContinentData from ContinentDict and label index and column
    ContinentCountries = pd.Series(dtype='str')
    for i, v in ContinentDict.items():
        ContinentCountries[v] = [i] if v not in ContinentCountries.keys() else ContinentCountries[v] + [i]

    ContinentData = pd.DataFrame(ContinentCountries)
    ContinentData.reset_index(inplace=True)
    ContinentData.columns =['Continent','ContinentCountries']

    population = populationEstimate()['popEst']

    continentPop  = []
    continentSize = []
    continentSum  = []
    continentMean = []
    continentStd  = []
    for continent in ContinentData['ContinentCountries']:
        pop = []
        for country in continent:
            pop.append(population[country].astype(int))
        continentPop.append(pop)
        continentSize.append(len(pop))
        continentSum.append(sum(pop))
        continentMean.append(np.mean(pop))
        continentStd.append(np.std(pop))

    ContinentData['Population'] = continentPop    
    ContinentData['size'] = continentSize    
    ContinentData['sum'] = continentSum    
    ContinentData['mean'] = continentSum    
    ContinentData['std'] = continentSum    

    ContinentData.index = ContinentData['Continent']

    ContinentData.drop(['Continent'], axis=1, inplace=True)
    ContinentData.drop(['ContinentCountries'], axis=1, inplace=True)
    ContinentData.drop(['Population'], axis=1, inplace=True)

    return ContinentData
    raise NotImplementedError()
    
````
