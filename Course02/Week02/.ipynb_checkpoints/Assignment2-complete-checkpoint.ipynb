```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime as dt
import mplleaflet
import pandas as pd
import numpy as np


def getStationData(binsize, hashid) :
    stationData = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))
    station_locations_by_hash = stationData[stationData['hash'] == hashid]

    return station_locations_by_hash


def getTemperatureData(binsize, hashid) :
    temperatureData = pd.read_csv('data/C2A2_data/BinnedCsvs_d{}/{}.csv'.format(binsize, hashid))

    temperatureData              = temperatureData.sort_values(by="Date")
    temperatureData['Date']      = pd.to_datetime(temperatureData['Date'])
    temperatureData['DayOfYear'] = pd.to_datetime(temperatureData['Date']).dt.dayofyear
    temperatureData              = temperatureData[~((temperatureData.Date.dt.month == 2) & \
                                                  (temperatureData.Date.dt.day == 29))] 
    temperatureData['AdjTemp']   = temperatureData['Data_Value'].apply(lambda T : float(T)/10.0)

    colSet = ["ID", "Date", "Element", "Data_Value", "DayOfYear", "AdjTemp"]

    tempLineData    = temperatureData.loc[temperatureData['Date'] < "2015-01-01",colSet]
    tempScatterData = temperatureData.loc[temperatureData['Date'] >= "2015-01-01",colSet]

    return [tempLineData, tempScatterData]


def getLinePointList(linePointData) :
    lowTempData  = linePointData[linePointData["Element"] == "TMIN"]
    highTempData = linePointData[linePointData["Element"] == "TMAX"]

    maxTempList = []
    minTempList = []
    dateList    = []

    for day in range(1,366) :
        dayLow   = lowTempData[lowTempData["DayOfYear"] == day]
        dayHigh  = highTempData[highTempData["DayOfYear"] == day]

        minTempList.append(dayLow['AdjTemp'].min())
        maxTempList.append(dayHigh['AdjTemp'].max())

        dateList.append(day)

    return [minTempList, maxTempList, dateList]


def getScatterPointList(minTempList, maxTempList, lineScatterData) :
    minScatterData  = pd.DataFrame(columns=["minDay", "minValue"])
    maxScatterData  = pd.DataFrame(columns=["maxDay", "maxValue"])

    lowTempData  = lineScatterData[lineScatterData["Element"] == "TMIN"]
    highTempData = lineScatterData[lineScatterData["Element"] == "TMAX"]

    for day in range(1, 365) :
        dayLow   = lowTempData[lowTempData["DayOfYear"] == day]["AdjTemp"]
        dayHigh  = highTempData[highTempData["DayOfYear"] == day]["AdjTemp"]

        if dayLow.min() < minTempList[day] :
            minScatterData.loc[len(minScatterData.index)] = [day, dayLow.min()]
        if dayHigh.max() > maxTempList[day] :
            maxScatterData.loc[len(maxScatterData.index)] = [day, dayHigh.max()]

    return [minScatterData, maxScatterData]


def createGraphFrame(ax, dateList) :
    locator    = mdates.DayLocator(bymonthday=(1, 365))

    ax.xaxis.set_major_locator(locator)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%B'))
    ax.set_xlim(min(dateList), max(dateList))

    ticklabels = ["Jan 01", "Feb 01", "Mar 01", "Apr 01", "May 01", "Jun 01", "Jul 01", "Aug 01", "Sep 01", "Oct 01", "Nov 01", "Dec 01"]
    ax.set_xticklabels(ticklabels)
    ax.set_xlabel('Day')
    ax.set_ylabel('Temperature ($^\circ$C)')
    plt.xticks(rotation=45)

def createLinePlot(minPointList, maxPointList, dateList, ax) :
    plt.ylim([-40, 45])

    ax.plot(dateList, maxPointList, color='red', linestyle=':', linewidth=0.5, label = "Max. Average Temp")
    ax.plot(dateList, minPointList, color='blue', linestyle='--', linewidth=0.3, label = "Min. Average Temp")


def createScatter(minScatterData, maxScatterData, ax) :
    ax.scatter(maxScatterData["maxDay"], maxScatterData["maxValue"], c = 'red', marker = 'x')
    ax.scatter(minScatterData["minDay"], minScatterData["minValue"], c = 'blue', marker = '+')


def leaflet_plot_stations(binsize, hashid):
    stationData                       = getStationData(binsize, hashid)
    [linePointData, lineScatterData]  = getTemperatureData(binsize, hashid)

    [minPointList, maxPointList, datePointList] = getLinePointList(linePointData)
    [minScatterData, maxScatterData]            = getScatterPointList(minPointList, maxPointList, lineScatterData)
    plt.rc('hatch', color='thistle', linewidth=0.5)
    fig = plt.figure(num=1, figsize=[8.0, 8.0])
    ax = fig.add_subplot(1, 1, 1)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    createGraphFrame(ax, datePointList)
    createLinePlot(minPointList, maxPointList, datePointList, ax)
    createScatter(minScatterData, maxScatterData, ax)

    plt.title("Daily averge temperatures extremes (2005-2014) and\ndays in 2015 exceeding those extremes", loc='center')
    plt.legend(('Avg. High Temp (2005-2014)', 'Avg. Low Temp (2005-2014)', '2015 Temps Above 2005-2014 High Average', '2015 Temps Below 2005-2014 Low Average'), loc='lower right', shadow=True)
    ax.fill_between(datePointList, minPointList, maxPointList, hatch = "x", color = 'thistle', alpha = 0.13)

    plt.savefig("temps.png")

    plt.show()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')


```
